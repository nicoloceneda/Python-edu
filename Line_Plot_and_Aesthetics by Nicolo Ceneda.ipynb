## Import Libraries

Import the Pyplot module from the Matplotlib library with the alias `plt` and NumPy with the
alias `np`.

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plot

Create a line plot.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Create a figure
fig = plt.figure()

# Create an axes
ax = plt.axes()

# Plot the lines
ax.plot(x, y1)
ax.plot(x, y2)
```

# Aesthetics

Matplotlib allows to customize the line color, line style, axes limits and labels.

## Line Color

Matplotlib allows to specify the color of a line plot. If a color is not specified, Matplotlib will cycle through a set of default colors. 

Create a line plot and set the line color.

```python
x = np.linspace(0, 10, 100)
```

```python
fig = plt.figure()
ax = plt.axes()

# Color by name
ax.plot(x, np.sin(x+1), color='blue')

# Color by short code
ax.plot(x, np.sin(x+2), color='g')

# Color by grayscale with values between 0 and 1
ax.plot(x, np.sin(x+3), color='0.75')

# Color by Hex code
ax.plot(x, np.sin(x+4), color='#FFDD44')

# Color by RGB tuple with values between 0 and 1
ax.plot(x, np.sin(x+5), color=(1.0,0.2,0.3))

# Color by HTML name
ax.plot(x, np.sin(x+5), color='chartreuse')
```

## Line Style

Matplotlib allows to specify the line style of a line plot.

Create a line plot and set the line style.

```python
x = np.linspace(0, 10, 100)
```

```python
fig = plt.figure()
ax = plt.axes()

# Solid 
ax.plot(x, np.sin(x+1), linestyle='solid') # Or: -

# Dashed
ax.plot(x, np.sin(x+2), linestyle='dashed') # Or: --

# Dash-dot
ax.plot(x, np.sin(x+3), linestyle='dashdot') # Or: -.

# Dotted
ax.plot(x, np.sin(x+4), linestyle='dotted') # Or: :
```

## Axes Limits

Matplotlib chooses automatically the limits of a plot but also allows to specify them manually. 

Create a line plot and set the limits of the axes.

```python
x = np.linspace(0, 10, 100)
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))
ax.set_xlim(-2.5, 12.5)
ax.set_ylim(-1.5, 1.5)
```

Note that the axes can also be inverted.

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))
ax.set_xlim(12.5, -2.5)
ax.set_ylim(1.5, -1.5)
```

When using the MATLAB interface, a useful method is `plt.axis()`, written with the 'i' not 'e'.

It allows to specify the x and y axes limits in one command.

```python
plt.plot(x, np.sin(x))

plt.axis([-2.5, 12.5, -1.5, 1.5])
```

It allows to automatically tighten the axes around the line plot.

```python
plt.plot(x, np.sin(x))

plt.axis('tight')
```

It allows to specify an equal aspect ratio so that one unit in x is equal to one unit in y on the screen.

```python
plt.plot(x, np.sin(x))

plt.axis('equal')
```

## Plot Labels

Matplotlib allows to specify titles, axis labels, and simple legends.

Create a line plot and set title and axis labels.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, y)
ax.set_title('sin function')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
```

Create a line plot and set a legend.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, y1, label='sin(x)')
ax.plot(x, y2, label='cos(x)')

# Display the legend
plt.legend()
```

## Set all aesthetic specs at once

The object oriented interface allows to set all properties at once.

Create a line plot and set all properties at once.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, y)
ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel='x', ylabel='sin(x)', title='A Simple Plot');
```

## MATLAB Interface

The following commands are equivalent to those mentioned above for the MATLAB interface:
* ax.set_xlabel == plt.xlabel
* ax.set_ylabel == plt.ylabel
* ax.set_xlim == plt.xlim
* ax.set_ylim == plt.ylim
* ax.set_title == plt.title
