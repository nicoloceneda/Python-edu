## Import Libraries

Import the Pyplot module from the Matplotlib library with the alias `plt` and NumPy with the alias `np`.

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Interfaces

Matplotlib provides two different interfaces: a MATLAB-style interface and an object-oriented interface.

## MATLAB Interface

The MATLAB interface keeps track of the current figure and axes, which are where all `plt` commands are applied. 

Create a figure with two subplots.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Create a figure
plt.figure()

# Create the first subplot (rows, columns, subplot number)
plt.subplot(2, 1, 1)
plt.plot(x, y1)

# Create the first subplot (rows, columns, subplot number)
plt.subplot(2, 1, 2)
plt.plot(x, y2)
```

## Object-Oriented Interface

Rather than depending on a notion of "active" figure or axis, the object-oriented interface uses the object methods `Figure` and `Axis`. 

Create a figure with two subplots.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Create a figure with two subplots
fig, ax = plt.subplots(2)

# Plot the first image
ax[0].plot(x, y1)

# Plot the second image
ax[1].plot(x, y2)
```

# Save Plot

The command `savefig()` allows to save the figure in the current working directory. Many different file formats are available and can be specified by the extension given to the file name. 

```python
fig.savefig('my_figure.png')
```

# Show Plot

When using Mtaplotlib from a script, the function `plt.show()` starts an event loop which looks for all active figure objects and opens one or more interactive windows to display them. The command should be used only once per Python session and it is usually entered at the end of the script. Repeting the command can lead to backend-dependent behavior.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x, y)

plt.show()
```
