# Pandas - Index Creation and Operations


---



```
import pandas as pd
```

## Creating indexes

Create an index **from a list** using `pd.Index`

```
# Numerical index
index_a = pd.Index([1, 2, 3])
print(index_a)
```

```
# String index
index_a = pd.Index(["a", "b", "c"])
print(index_a)
```

## Manipulating indexes

Create an index from the **union** of indexes using `|`

```
index_a = pd.Index([1, 2, 3, 4])
index_b = pd.Index([1, 2, 5, 6])
index_c = index_a | index_b
print(index_c)
```

Create an index from the **intersection** of indexes using `&`

```
index_a = pd.Index([1, 2, 3, 4])
index_b = pd.Index([1, 2, 5, 6])
index_c = index_a & index_b
print(index_c)
```

Create an index from the **symmetric difference** of indexes using `^`

```
index_a = pd.Index([1, 2, 3, 4])
index_b = pd.Index([1, 2, 5, 6])
index_c = index_a ^ index_b
print(index_c)
```
