*IPython (Interactive Python) provides an interactive interface to Python, syntactic additions to the language, and it is closely tied to the Jupyter project, which provides a browser-based notebook that is useful for development, collaboration, sharing and publication of data science results. The IPython notebook is a special case of the broader Jupyter notebook structure, which encompasses notebooks for Julia, R and other programming languages.*



# Shell or Notebook

There are two primary means of using IPython: the IPython shell and the IPython notebook.

## Launching the Ipython Shell

**Launch the IPython shell** by typing **`ipython`** in the terminal.

```
nicolo@mac-nicolo ~ % ipython
Python 3.8.5 (v3.8.5:580fbb018f, Jul 20 2020, 12:11:27) 
Type 'copyright', 'credits' or 'license' for more information
IPython 7.17.0 -- An enhanced Interactive Python. Type '?' for help.
```


## Launching the Jupyter Notebook

The Jupyter notebook is a browser-based graphical interface to the IPython shell. Though the IPython notebook is viewed and edited through your web browser window, it must connect to a running Python process (known as kernel) in order to execute code. 

**Launch the Jupyter Notebook** by typing **`jupyter notebook`** in the terminal.

```
nicolo@mac-nicolo ~ % jupyter notebook
[I 01:56:24.672 NotebookApp] Serving notebooks from local directory: /Users/nicolo
[I 01:56:24.672 NotebookApp] Jupyter Notebook 6.1.3 is running at:
[I 01:56:24.672 NotebookApp] http://localhost:8888/...
[I 01:56:24.672 NotebookApp]  or http://127.0.0.1:8888/...
[I 01:56:24.672 NotebookApp] Use Control-C to stop this server and shut down all kernels (twice to skip confirmation).
[C 01:56:24.677 NotebookApp] 
    
    To access the notebook, open this file in a browser:
        file:///Users/nicolo/Library/Jupyter/runtime/nbserver-13423-open.html
    Or copy and paste one of these URLs:
        http://localhost:8888/...
     or http://127.0.0.1:8888/...
```

Upon issuing the command, your default browser should automatically open and navigate to the listed local URL; the exact address will depend on your system. If the browser does not open automatically, you can open a window and manually open one of the listed addresses.
